##### Introduction

My project aims to analyze how effective the CDC's Social Vulnerability Index (SVI) is in
representing Native Hawaiian or Pacific Islander (NHPI) populations, which have been historically
underrepresented in data. I will be comparing the highest risk quartile areas based on the SVI with
Census NHPI demographic data from either the Decennial Census or American Community Survey
tables on a statewide and county level in California (the state with the highest NHPI population after Hawaii). I was motivated to further research this topic from working in the NHPI Data Policy Lab and from community feedback.

The basic question I want to answer is if NHPI populations are underrepresented in the SVI (given that they don't explicitly factor NHPI populations in the index calculation) and to what extent. I want to further analyze the specific geographical areas wherein the largest disparities exist. 

Links to my data can be found at the following links: https://www.atsdr.cdc.gov/placeandhealth/svi/data_documentation_download.html
and https://data.census.gov/

##### About the Data

SVI data, which can be directly downloaded from the CDC's website and visualized in a map on the same website, is published every two years. The most recent data is from 2020. The SVI calculates a score for a myriad of indicators in the following categories: socioeconomic status, household characteristics, racial and ethnic minority status, and housing type and transporation. It then computes the percentile ranking of all indicators combined into an overall index between 0 and 1. 

I first wanted to look at a few variables in the dataset: RPL_THEME1, RPL_THEME2, RPL_THEME3, RPL_THEME4, and RPL_THEMES. The first four represent the individual scores for the indicator categories in respective order, and RPL_THEMES is the percentile ranking of all combined.

```python
# first import dataset
import pandas as pd
import matplotlib.pyplot as plt
svi_counties = pd.read_csv('California_COUNTY.csv')
```

Below are some summary statistics of the variables of interest in California counties. 

```python
svi_counties['RPL_THEME1'].describe()
```

```python
svi_counties['RPL_THEME2'].describe()
```

```python
svi_counties['RPL_THEME3'].describe()
```

```python
svi_counties['RPL_THEME4'].describe()
```

```python
svi_counties['RPL_THEMES'].describe()
```

All variables have similar summary statistics, which suggests that all four indicator variables factored in somewhat equally into the overall SVI score. However, a mean of exactly 0.5 in the RPL_THEMES variable (the overall SVI score) indicates that some standardization procedures were likely applied to create an equal distribution of index scores. 

Below is a simple histogram of the distribution SVI rankings on the county level in California to corroborate my hypotheses:

```python
svi_counties.plot(y='RPL_THEMES', kind='hist', bins=15)
plt.title("Distribution of SVI Index Across California Counties")
plt.xlabel("SVI Index Score")
plt.legend(["SVI Index Score"])
```

It looks like the distribution of SVI index scores across California counties is very symmetrical. However, this may be because the values themselves are small. I will clean the dataset so that the RPL_THEMES variable (the overall index score) is reformatted into a percentage. I will make a new column in the svi_counties dataframe.

```python
svi_counties['RPL_Percent'] = svi_counties['RPL_THEMES']*100
```

Now time to replot the histogram distribution of index scores:

```python
svi_counties.plot(y='RPL_Percent', kind='hist', bins=10)
plt.title("Distribution of SVI Index Across California Counties")
plt.xlabel("SVI Index Score")
plt.legend(["SVI Index Score"])
```

The plots look the same. As a result, I can conclude that no significant errors exist in my data and that the SVI index scores are equally distributed across counties.

Next, I will check the distribution of SVI index scores across California Census tracts, which have arguably more variation.

```python
svi_tracts = pd.read_csv('California_census.csv') # read in dataframe from CSV
```

```python
svi_tracts.plot(y='RPL_THEMES', kind='hist')
plt.title("Distribution of SVI Index Across California Census Tracts")
plt.xlabel("SVI Index Score")
plt.legend(["SVI Index Score"])
```

The histogram immediately clues me in that there are significant outliers in the data that don't make sense (since index scores are only from 0 to 1) and are skewing the histogram. Upon further inspection, I found that some values for certain Census tracts were labeled as '-999', indicating that no data existed for that Census tract specifically. As a result, I decided to filter out any rows containing -999 for now.

```python
# create new filtered dataset
svi_tracts_filtered = svi_tracts[(svi_tracts['RPL_THEMES'] != -999)]
```

```python
svi_tracts_filtered.plot(y='RPL_THEMES', kind='hist')
plt.title("Distribution of SVI Index Across California Census Tracts")
plt.xlabel("SVI Index Score")
plt.legend(["SVI Index Score"])
```

My updated histogram that filtered out the -999 values now shows almost a completely even distribution of SVI index scores across Census tracts in California. This is consistent with my county level results and is useful as to determining how SVI index scores were calculated.

##### Data Analysis

Now let's take a look at the distribution of NHPI populations in California. I'll first look at Los Angeles County and its Census tracts, as it has the largest population of NHPIs in California.

```python
# import geopandas to read geoJSON file
import geopandas as gpd
la_nhpi_pop = gpd.read_file('acs2021_5yr_B02012_14000US06037218800.geojson')
```

```python
la_nhpi_pop.head()
```

From taking a quick preview of the geopandas dataframe, the variable B02012001 denotes the NHPI population. I will now visualize this in map format.

```python
la_nhpi_pop.plot(figsize=(12,10),
         column='B02012001',
         legend=True,
         scheme='NaturalBreaks')

plt.title('Distribution of NHPI Populations in Los Angeles County by Census Tract Level',fontsize=16)
```

I will now focus on combining the datasets for SVI index scores from CDC data and NHPI populations from Census data.

Let's start again with Los Angeles County. The geopandas dataframe from Census data had a geoid that corresponds to the FIPS code of SVI index score data. I will aim to combine the two dataframes based on this column.

```python
# first reformat geoid so that it corresponds to SVI data 
la_nhpi_pop['FIPS'] = la_nhpi_pop['geoid'].str[-10:]
```

```python
la_nhpi_pop.head()
```

Now filter svi_tracts_filtered for Los Angeles County.

```python
la_tracts_filtered = svi_tracts_filtered[svi_tracts_filtered['COUNTY'] == 'Los Angeles']
```

```python
la_nhpi_pop.at[0, 'FIPS'] = '06037'
```

```python
la_nhpi_pop['FIPS']=la_nhpi_pop['FIPS'].astype(int)
```

Merge the two dataframes together.

```python
merged_la = pd.merge(la_nhpi_pop, la_tracts_filtered, on='FIPS')
```

```python
merged_la.plot(figsize=(12,10),
         column='RPL_THEMES',
         legend=True,
         scheme='NaturalBreaks')

plt.title('Distribution of SVI Index Scores in Los Angeles County at theb Census Tract Level',fontsize=16)
```

Next step: change the color gradient of each map so that both NHPI and SVI index score color gradients correspond with each other.

Looking at the overall maps of SVI index score and NHPI populations for LA County Census tracts, there seem to be obvious discrepancies between areas with higher index scores (indicating higher vulnerability) and with higher NHPI populations. For instance, some counties on the west side of LA County have high NHPI populations but lower SVI index scores overall.
